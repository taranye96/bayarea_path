```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Read in velocity values
velmod_file = np.genfromtxt('/Users/tnye/bayarea_path/files/velmod/bay_3Dvelmod_utm-cartesian.out')

# Extract coordinates and velocity values
utm_x = velmod_file[:,0]
utm_y = velmod_file[:,1]
depth = velmod_file[:,2]
Vp = velmod_file[:,4]
Vs = velmod_file[:,5]
```

```python
print(np.max(Vp))
print(np.max(Vs))
```

```python
# Plot Vp velocity model
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')

sc = ax.scatter(utm_x, utm_y, depth, c=Vp/1000, cmap='plasma', marker='.', s=0.5, vmin=0)

# Add color bar
mappable = plt.cm.ScalarMappable(cmap='plasma')
mappable.set_array(Vp)
mappable.set_clim(0, 8)
cbar = fig.colorbar(mappable, ax=ax, shrink=0.5, aspect=5)
cbar.set_label('Vp (km/s)')

# Set labels
ax.set_xlabel('UTMx')
ax.set_ylabel('UTMy')
ax.set_zlabel('Depth (m)')
ax.set_title('Bay Area Vp Velocity Model')

# Show the plot
plt.show()
```
